```python
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
# import seaborn as sns

G=nx.DiGraph()
G.add_node('Bashful')
G.add_node('Doc')
G.add_node('Dopey')
G.add_node('Happy')
G.add_node('Sleepy')
G.add_node('Sneezy ')
G.add_node('Grumpy')
list(G.nodes())

G.add_edge('Bashful', 'Doc')
G.add_edge('Doc', 'Dopey')
G.add_edge('Dopey', 'Happy')
G.add_edge('Happy', 'Sleepy')
G.add_edge('Sleepy', 'Sneezy ')
G.add_edge('Sneezy ', 'Grumpy')
G.add_edge('Grumpy', 'Bashful')

list(G.edges())

nx.draw_circular(G,
                 node_size=2000,
                 with_labels=True)
plt.axis('equal')
```

```python
positions =dict(Vienna=(48, 16),
               Alexandria=(31, 30),
                Brasilia=(-15, -47),
                Kyiv=(50, 30),
               Yerevan=(40, 44),
               Rome=(41, 12),
               Busan=(35, 129),
               Saint_Petersburg=(60, 30),
               Minsk=(53, 27),
               Algiers=(36, 3))
positions['Vienna']
```

```python
G = nx.Graph()
G.add_nodes_from(positions)
G.nodes()
```

```python
drive_times = {('Vienna', 'Alexandria'):3,
              ('Alexandria', 'Brasilia'):12,
              ('Brasilia', 'Kyiv'):13,
              ('Kyiv', 'Alexandria'):3,
              ('Kyiv', 'Yerevan'):2,
              ('Kyiv', 'Rome'):2,
              ('Kyiv', 'Busan'):10,
              ('Kyiv', 'Saint_Petersburg'):1,
              ('Kyiv', 'Minsk'):1,
              ('Kyiv', 'Algiers'):3}

G.add_edges_from(drive_times)
G.edges()

nx.draw(G,
       positions,
       node_color='C1',
       node_shape='s',
       node_size=2000,
       with_labels=True)

nx.draw_networkx_edge_labels(G, 
                             positions, 
                            edge_labels=drive_times)
# plt.axis('equal')
```

```python
shortest_path_distance = nx.dijkstra_path(G, 'Vienna', 'Alexandria')
shortest_path_distance
```
